# üè• Healthcare Dataset Cleaning & Analysis

### This script performs data cleaning, validation, and basic exploratory analysis on a healthcare dataset using Pandas and Matplotlib.

### üì¶ Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### üìÅ Load Dataset

```python
df = pd.read_csv('healthcare_dataset.csv')
df
```

###  üßπ Clean Column Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df
```

### üî§ Standardize Text Fields

```python
df['name'] = df['name'].str.title().str.strip()
df['doctor'] = df['doctor'].str.title().str.strip()
df['hospital'] = df['hospital'].str.title().str.strip()
df
```

```python
df['gender'] = df['gender'].str.capitalize().str.strip()
```

### üìÖ Convert Date Columns

```python
df['date_of_admission'] = pd.to_datetime(df['date_of_admission'], errors='coerce')
df['discharge_date'] = pd.to_datetime(df['discharge_date'], errors='coerce')
df
```

### üßæ Standardize Additional Text Columns

```python
text_columns = ['medical_condition', 'admission_type', 'medication', 'test_results']
for col in text_columns:
    df[col] = df[col].str.title().str.strip()
```


### ‚ùì Handle Missing Data

```python
missing = df.isnull().sum()
print(missing)

# Example: Fill missing room_number with median (if needed)
df['room_number'] = df['room_number'].fillna(df['room_number'].median())
df
```

### üî¢ Convert Numeric Fields

```python
df['billing_amount'] = pd.to_numeric(df['billing_amount'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['room_number'] = pd.to_numeric(df['room_number'], errors='coerce')
```

### üè® Calculate Length of Stay

```python
df['length_of_stay'] = (pd.to_datetime(df['discharge_date']) - pd.to_datetime(df['date_of_admission'])).dt.days
```

### üöª Validate Gender Values

```python
df['gender'] = df['gender'].str.strip().str.title()
valid_genders = ['Male', 'Female', 'Other']
df = df[df['gender'].isin(valid_genders)]
```

### ü©∏ Validate Blood Type


```python
valid_blood_types = ['A+', 'A-', 'B+', 'B-', 'AB+', 'AB-', 'O+', 'O-']
df['blood_type'] = df['blood_type'].str.strip().str.upper()
df = df[df['blood_type'].isin(valid_blood_types)]
```

### üóì Standardize Date Formats

```python
df['date_of_admission'] = pd.to_datetime(df['date_of_admission'], format='%d-%m-%Y', errors='coerce')
df['discharge_date'] = pd.to_datetime(df['discharge_date'], format='%d-%m-%Y', errors='coerce')

# Drop rows with invalid dates
df = df.dropna(subset=['date_of_admission', 'discharge_date'])

# Remove rows where discharge is before admission
df = df[df['discharge_date'] >= df['date_of_admission']]
```

### üßÆ Fix Length of Stay Inconsistencies

```python
df['calculated_stay'] = (df['discharge_date'] - df['date_of_admission']).dt.days

# Find mismatched length_of_stay
stay_mismatch = df[df['length_of_stay'] != df['calculated_stay']]

# Show incorrect rows
stay_mismatch[['name', 'length_of_stay', 'calculated_stay']]
```

```python
# Replace original with correct value
df['length_of_stay'] = df['calculated_stay']
df.drop(columns=['calculated_stay'], inplace=True)
```

### üßë‚Äç‚öïÔ∏è Remove Titles from Names & Doctors

```python
df['doctor'] = df['doctor'].str.replace(r'(Dr\.|Mr\.|Mrs\.|Ms\.)', '', regex=True).str.strip()
df['name'] = df['name'].str.replace(r'(Dr\.|Mr\.|Mrs\.|Ms\.)', '', regex=True).str.strip()
```

### üè• Handle Commas in Hospital Names


```python
df['hospital'] = df['hospital'].apply(lambda x: f'"{x}"' if ',' in x else x)
```

### üíæ Save Cleaned Data

```python
df.to_csv("cleaned_medical_data.csv", index=False)
```
